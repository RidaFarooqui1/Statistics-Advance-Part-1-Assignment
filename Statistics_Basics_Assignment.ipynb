* 10.Explain nominal, ordinal, interval, and ratio levels of measurement.

Ans->

* Nominal:

 This is the most basic level. Data is categorized into groups, but there's no inherent order or ranking. Examples include hair color (blonde, brown, black), gender (male, female), or types of vehicles (car, truck, motorcycle).
* Ordinal:

  This level allows for categorization and ranking, but the differences between categories may not be equal or quantifiable. Examples include grades (A, B, C, D, F), satisfaction levels (very dissatisfied, dissatisfied, neutral, satisfied, very satisfied), or rankings in a competition.

 This level allows for categorization, ranking, and measurement with equal intervals. However, there's no true zero point, meaning that the zero value doesn't indicate the absence of the variable. Examples include temperature measured in Celsius or Fahrenheit, or dates on a calendar.
* Ratio:

 This is the highest level of measurement. It includes all features of interval data plus a true zero point, meaning that zero represents the absence of the variable. Examples include weight, height, age, or income.

* 2.Write a Python program to compute the variance and standard deviation of a dataset.

````python
import numpy as np

def calculate_variance_and_std(data):
    """
    Calculates the variance and standard deviation of a dataset.

    Args:
      data: A list or numpy array of numerical data.

    Returns:
        A tuple containing the variance and standard deviation.
    """
    if not isinstance(data, (list, np.ndarray)):
        raise TypeError("Input data must be a list or numpy array.")
    if len(data) < 2:
        raise ValueError("Data must contain at least two elements.")

    data = np.array(data)
    variance = np.var(data)
    standard_deviation = np.std(data)
    return variance, standard_deviation


if __name__ == "__main__":
    dataset = [1, 2, 3, 4, 5]
    variance, std_dev = calculate_variance_and_std(dataset)
    print(f"Variance: {variance}")
    print(f"Standard Deviation: {std_dev}")
````

* 4.Implement sampling techniques like random sampling and stratified sampling.

````python
# Random sampling
import random

def random_sampling(population, sample_size):
    """
    Selects a random sample from a population.

    Args:
        population: A list representing the population.
        sample_size: The desired size of the sample.

    Returns:
        A list containing the random sample.
    """
    if sample_size > len(population):
        raise ValueError("Sample size cannot be greater than population size.")
    return random.sample(population, sample_size)


population = list(range(1, 101))
sample_size = 20
sample = random_sampling(population, sample_size)
print("Random sample:", sample)
````

````python
# Stratified Sampling
import pandas as pd

def stratified_sampling(df, strata_column, sample_size_per_stratum):
    """
    Performs stratified sampling on a DataFrame.

    Args:
        df: The DataFrame containing the population data.
        strata_column: The name of the column used to define strata.
        sample_size_per_stratum: The desired sample size for each stratum.

    Returns:
        A DataFrame containing the stratified sample.
    """
    if not isinstance(df, pd.DataFrame):
        raise TypeError("Input must be a pandas DataFrame.")
    if strata_column not in df.columns:
        raise ValueError(f"Column '{strata_column}' not found in DataFrame.")
    return df.groupby(strata_column, group_keys=False).apply(lambda x: x.sample(min(len(x), sample_size_per_stratum)))


data = {'group': ['A', 'A', 'A', 'B', 'B', 'C', 'C', 'C', 'C'],
        'value': [10, 12, 15, 20, 22, 30, 35, 32, 31]}
df = pd.DataFrame(data)

sample_size_per_stratum = 2
stratified_sample = stratified_sampling(df, 'group', sample_size_per_stratum)
print("Stratified sample:\n", stratified_sample)
````

* 6.Create a dataset and plot its histogram to visualize skewness.

````python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

def generate_skewed_data(size, skewness):
    if skewness > 0:
        data = np.random.exponential(scale=1, size=size)
        data = data * (skewness/1.5)
    elif skewness < 0:
        data = -np.random.exponential(scale=1, size=size)
        data = data * (abs(skewness)/1.5)
    else:
        data = np.random.normal(size=size)
    return data

size = 1000
skewness = 2

data = generate_skewed_data(size, skewness)

plt.hist(data, bins=30, color='skyblue', edgecolor='black')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title(f'Histogram with Skewness = {skewness}')

skew_value = stats.skew(data)
plt.text(x=0.05, y=0.95, s=f"Skew: {skew_value:.2f}", transform=plt.gca().transAxes, verticalalignment='top')

plt.grid(axis='y', alpha=0.5)
plt.show()
````

* 9.Write a Python script to calculate covariance between two datasets.

````python
import numpy as np

def calculate_covariance(data1, data2):
    """Calculates the covariance between two datasets.

    Args:
        data1: A list or numpy array of numerical data.
        data2: A list or numpy array of numerical data.

    Returns:
        The covariance between the two datasets.
    """

    if len(data1) != len(data2):
        raise ValueError("Datasets must have the same length.")

    n = len(data1)
    mean1 = np.mean(data1)
    mean2 = np.mean(data2)

    covariance = np.sum((data1 - mean1) * (data2 - mean2)) / (n - 1)
    return covariance

if __name__ == '__main__':
    dataset1 = [1, 2, 3, 4, 5]
    dataset2 = [2, 4, 5, 4, 5]

    covariance = calculate_covariance(dataset1, dataset2)
print(f"The covariance between dataset1 and dataset2 is: {covariance}")
````

````python
import numpy as np
from scipy.stats import pearsonr

def calculate_correlation(dataset1, dataset2):
    """
    Calculates the Pearson correlation coefficient between two datasets.

    Args:
        dataset1 (list or numpy.ndarray): The first dataset.
        dataset2 (list or numpy.ndarray): The second dataset.

    Returns:
        float: The Pearson correlation coefficient, or None if an error occurs.
    """
    try:
        dataset1 = np.array(dataset1)
        dataset2 = np.array(dataset2)

        if len(dataset1) != len(dataset2):
            raise ValueError("Datasets must have the same length.")

        correlation, _ = pearsonr(dataset1, dataset2)
        return correlation
    except ValueError as e:
        print(f"Error: {e}")
        return None
    except Exception as e:
        print(f"An unexpected error occurred: {e}")
        return None


if __name__ == "__main__":
    data1 = [1, 2, 3, 4, 5]
    data2 = [2, 4, 5, 4, 5]

    correlation = calculate_correlation(data1, data2)

    if correlation is not None:
        print(f"The Pearson correlation coefficient is: {correlation}")

    data3 = [1, 2, 3]
    data4 = [4, 5, 6, 7]
    correlation = calculate_correlation(data3, data4)
````

* 12. Implement and compare simple random sampling and systematic sampling.

````python
#Simple Random Sampling
import random

def simple_random_sampling(population, sample_size):
    """
    Selects a random sample from a population.

    Args:
        population: A list representing the population.
        sample_size: The desired size of the sample.

    Returns:
        A list containing the random sample.
    """
    if sample_size > len(population):
        raise ValueError("Sample size cannot be greater than population size.")
    return random.sample(population, sample_size)

population = list(range(1, 101))
sample_size = 20
sample = simple_random_sampling(population, sample_size)
print("Simple Random Sample:", sample)
````

````python
#Systematic Sampling
import random

def systematic_sampling(population, sample_size):
    """
    Selects a systematic sample from a population.

    Args:
        population: A list representing the population.
        sample_size: The desired size of the sample.

    Returns:
        A list containing the systematic sample.
    """
    population_size = len(population)
    if sample_size > population_size:
        raise ValueError("Sample size cannot be greater than population size.")

    interval = population_size // sample_size
    start_index = random.randint(0, interval - 1)
    sample = [population[i] for i in range(start_index, population_size, interval)]

    return sample[:sample_size]

population = list(range(1, 101))
sample_size = 20
sample = systematic_sampling(population, sample_size)
print("Systematic Sample:", sample)
````

* 13.Calculate the mean, median, and mode of grouped data.

````python
# Mean
import pandas as pd

def calculate_grouped_mean(data):
  """Calculates the mean of grouped data.

    Args:
        data: A pandas DataFrame with columns for class intervals (e.g., 'interval')
        and their frequencies (e.g., 'frequency').

    Returns:
        The mean of the grouped data.
  """
  data['midpoint'] = (data['interval'].apply(lambda x: int(x.split('-')[0])) + data['interval'].apply(lambda x: int(x.split('-')[1]))) / 2
  data['fx'] = data['midpoint'] * data['frequency']
  mean = data['fx'].sum() / data['frequency'].sum()
  return mean

grouped_data = {'interval': ['0-10', '10-20', '20-30', '30-40'],
                'frequency': [5, 15, 10, 8]}
grouped_df = pd.DataFrame(grouped_data)

mean = calculate_grouped_mean(grouped_df)
print(f"The mean of the grouped data is: {mean}")
````

````python
# Median
import pandas as pd

def calculate_grouped_median(data):
  """Calculates the median of grouped data.

    Args:
        data: A pandas DataFrame with columns for class intervals (e.g., 'interval')
        and their frequencies (e.g., 'frequency').

    Returns:
        The median of the grouped data.
  """
  data['cumulative_frequency'] = data['frequency'].cumsum()
  N = data['frequency'].sum()
  median_class = data[data['cumulative_frequency'] >= N/2].iloc[0]
  L = int(median_class['interval'].split('-')[0])
  CF = data[data['cumulative_frequency'] < median_class['cumulative_frequency']]['cumulative_frequency'].iloc[-1] if not data[data['cumulative_frequency'] < median_class['cumulative_frequency']].empty else 0
  f = median_class['frequency']
  h = int(median_class['interval'].split('-')[1]) - L
  median = L + ((N/2 - CF) / f) * h
  return median
print(calculate_grouped_median(grouped_df))
````

````python
# Mode
import pandas as pd

def calculate_grouped_mode(data):
  """Calculates the mode of grouped data.

    Args:
        data: A pandas DataFrame with columns for class intervals (e g., 'interval')
        and their frequencies (e.g., 'frequency').

    Returns:
        The mode of the grouped data.
  """
  modal_class_index = data['frequency'].idxmax()
  modal_class = data.loc[modal_class_index]
  L = int(modal_class['interval'].split('-')[0])
  f1 = modal_class['frequency']
  f0 = data.loc[modal_class_index-1, 'frequency'] if modal_class_index > 0 else 0
  f2 = data.loc[modal_class_index+1, 'frequency'] if modal_class_index < len(data)-1 else 0
  h = int(modal_class['interval'].split('-')[1]) - L
  mode = L + ((f1 - f0) / (2*f1 - f0 - f2)) * h
  return mode
print(calculate_grouped_mode(grouped_df))
````

* 18.Implement Z-score normalization and explain its significance.

````python
import numpy as np

def z_score_normalization(data):
    """
    Performs Z-score normalization on a given dataset.

    Args:
        data (np.array): The input data.

    Returns:
        np.array: The normalized data.
    """
    mean = np.mean(data)
    std_dev = np.std(data)
    normalized_data = (data - mean) / std_dev
    return normalized_data

data = np.array([1, 2, 3, 4, 5])
normalized_data = z_score_normalization(data)
print("Original Data:", data)
print("Normalized Data:", normalized_data)

# its signifanceZ-score normalization, also known as standardization,
#is a data preprocessing technique used to rescale numerical features to have a mean of 0 and a standard deviation of 1.
#This process transforms the data into a standard normal distribution, which is beneficial for many machine-learning algorithms.
````

* 22.Generate a dataset and implement both variance and standard deviation computations.

````python
import math

def calculate_variance(data):
    n = len(data)
    if n == 0:
        return 0
    mean = sum(data) / n
    squared_differences = [(x - mean) ** 2 for x in data]
    variance = sum(squared_differences) / n
    return variance

def calculate_std_dev(data):
    variance = calculate_variance(data)
    std_dev = math.sqrt(variance)
    return std_dev

data = [10, 12, 23, 23, 16, 23, 21, 16]

variance = calculate_variance(data)
print("Variance:", variance)

std_dev = calculate_std_dev(data)
print("Standard Deviation:", std_dev)
````

````python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
from scipy.stats import skew, kurtosis

np.random.seed(0)
data_normal = np.random.normal(0, 1, 1000)
data_skewed_pos = np.random.exponential(1, 1000)
data_skewed_neg = -np.random.exponential(1, 1000)
data_leptokurtic = np.random.laplace(0, 1, 1000)
data_platykurtic = np.random.uniform(-2, 2, 1000)

df = pd.DataFrame({
    'Normal': data_normal,
    'Skewed_Positive': data_skewed_pos,
    'Skewed_Negative': data_skewed_neg,
    'Leptokurtic': data_leptokurtic,
    'Platykurtic': data_platykurtic
})

def plot_distribution(data, title):
    plt.figure(figsize=(8, 6))
    sns.histplot(data, kde=True, color='skyblue')

    skewness_val = skew(data)
    kurtosis_val = kurtosis(data)

    plt.title(f'{title} Distribution\nSkewness: {skewness_val:.2f}, Kurtosis: {kurtosis_val:.2f}')
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.grid(axis='y', alpha=0.75)
    plt.show()

for column in df.columns:
    plot_distribution(df[column], column)
````

* 24.Implement the Pearson and Spearman correlation coefficients for a dataset.

````python
import numpy as np
from scipy.stats import pearsonr, spearmanr

def calculate_correlations(data_x, data_y):
    """
    Calculates Pearson and Spearman correlation coefficients for two datasets.

    Args:
      data_x: A list or numpy array of numerical data for the first variable.
      data_y: A list or numpy array of numerical data for the second variable.

    Returns:
      A tuple containing:
        - Pearson correlation coefficient (float)
        - Pearson correlation p-value (float)
        - Spearman correlation coefficient (float)
        - Spearman correlation p-value (float)
    """
    pearson_corr, pearson_p_value = pearsonr(data_x, data_y)
    spearman_corr, spearman_p_value = spearmanr(data_x, data_y)

    return pearson_corr, pearson_p_value, spearman_corr, spearman_p_value

data_x = np.array([1, 2, 3, 4, 5])
data_y = np.array([2, 4, 5, 4, 5])

pearson_corr, pearson_p_value, spearman_corr, spearman_p_value = calculate_correlations(data_x, data_y)

print(f"Pearson Correlation: {pearson_corr:.3f} (p-value: {pearson_p_value:.3f})")
print(f"Spearman Correlation: {spearman_corr:.3f} (p-value: {spearman_p_value:.3f})")
````
